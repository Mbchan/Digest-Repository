### Log Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.multiclass import OneVsRestClassifier

from sklearn.metrics import classification_report
```

```python
X = df_new['lemma_separated']
y = df_new.drop(columns=['lemma_separated'], axis=1)

# split dataset into training and validation set
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=46)

print('TRAIN')
display(X_train)
print('TEST')
display(X_test)
```

```python
tfidf_vectorizer = TfidfVectorizer(strip_accents='unicode', ngram_range=(1,3))

# create TF-IDF features
xtrain_tfidf = tfidf_vectorizer.fit_transform(X_train)
xtest_tfidf = tfidf_vectorizer.transform(X_test)
```

```python

categories =['chastity', 'civil_status', 'drugs',\
            'fundamental_laws', 'honor', 'liberty', 'national_security',\
            'persons', 'property', 'public_interest', 'public_morals',\
            'public_officers', 'public_order', 'quasi_offense']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier

# Using pipeline for applying logistic regression and one vs rest classifier
LogReg_pipeline = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag', random_state=42), n_jobs=-1)),
            ])

final_pred = pd.DataFrame()

for category in categories:
    print('**Processing {} cases...**'.format(category))
    
    # Training logistic regression model on train data
    LogReg_pipeline.fit(xtrain_tfidf, y_train[category])
    
    # calculating test accuracy
    prediction = LogReg_pipeline.predict(xtest_tfidf)
    print('Test accuracy is {}'.format(accuracy_score(y_test[category], prediction)))
    result = pd.DataFrame(prediction).rename(columns={0:category})
#     display(result)
    final_pred = pd.concat([final_pred, result], axis=1)
#     display(final_pred)
    print(classification_report(y_test[category], prediction))
    print("\n")
```
